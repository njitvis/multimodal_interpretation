```python
import pandas as pd
url = 'https://drive.google.com/file/d/1IAEH0-uwn48GTh_QArnUi8S67DfDrB5c/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df
```

```python
import plotly.express as px

fig = px.histogram(
    df.distance,
    nbins=30,
    template='simple_white',
    title="Distance Distribution (chart+caption)",
    width=1000,
    text_auto=True  # Display count as text on bars
).update_layout(title_x=0.5, showlegend=False, xaxis_title='Distance')

fig.update_traces(marker_color='gray')
fig.show()
```
